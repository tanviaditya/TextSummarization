The Facebook BART Transformer model is an approach to machine learning which combines two existing algorithms, BART and Transformer.

BART is a model composed of a mixture of classic techniques and it is a model focused on regression. Conversely, Transformer is a recurrent model which does not use parameter optimization.

```python
!pip install transformers
```

```python
from transformers import pipeline
smr_bart = pipeline(task="summarization", model="facebook/bart-large-cnn")
smbart = smr_bart(text, max_length=130, min_length=30, do_sample=False)
```

Summarized Text

```python
print(smbart[0]['summary_text'])
```

# Using the pre-trained model on fake news dataset

Data importing and Data cleaning

```python
import pandas as pd
```

```python
#Get data
df = pd.read_csv('/content/news_articles.csv',error_bad_lines=False, sep=',')
df.head(3)
```

```python
df = df.dropna().reset_index()
```

```python
#Select part of data we want to keep
df = df[(df['language']=='english') & (df['type']=='bs')].reset_index()
df = df[['title','text']]
```

```python
df.head()
```

```python
df.shape
```

```python
#Clean text
df['text'] = df['text'].apply(lambda x: str(x).replace('\n',''))

#Select only part of it (makes testing faster)
articles = df.head(5)
articles.head()
```

```python
from transformers import pipeline
smr_bart = pipeline(task="summarization", model="facebook/bart-large-cnn")
```

```python
new_df = pd.DataFrame(columns=['Text','Summary'])
new_df['Summary'] = articles['text'].apply(lambda x: smr_bart(x, max_length=130, min_length=30, do_sample=False)[0]['summary_text'])
new_df.head(3)
```

```python
new_df['Text']=articles['text']
```
